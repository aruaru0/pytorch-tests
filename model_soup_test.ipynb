```python
def copy_params(model1, model2) :
  model1.load_state_dict(model2.state_dict())
  return model1

def sum_model_params(modelA, modelB):
    """ modelA + modelB """
    sdA = modelA.state_dict()
    sdB = modelB.state_dict()
    for key in sdA:
        sdA[key] = (sdA[key] + sdB[key])
    modelA.load_state_dict(sdA)
    return modelA

def multi_model_params(model, a):
    """ a * model """
    sd = model.state_dict()
    for key in sd:
        sd[key] = sd[key] * a
    model.load_state_dict(sd)
    return model
```

```python
model1 = timm.create_model("resnet18", pretrained=True)
model2 = timm.create_model("resnet18", pretrained=True)
```

```python
model = sum_model_params(model1, model2)
model = multi_model_params(model, 0.5)
# model.state_dict()
```
