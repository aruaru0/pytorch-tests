```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(-100, 100, 1000)
x[0] = 10_000
x_dash = (x - np.min(x))/(np.max(x) - np.min(x))

fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
ax1.hist(x)

ax2 = fig.add_subplot(1,2,2)
ax2.hist(x_dash)

plt.show()
```

```python
x = np.random.randint(-100, 100, 1000)
x[0] = 10_000
x_dash = (x - np.mean(x))/(np.std(x))

fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
ax1.hist(x)

ax2 = fig.add_subplot(1,2,2)
ax2.hist(x_dash)

plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn import datasets
import pandas as pd
import seaborn as sns


wine = datasets.load_wine()
df =  pd.DataFrame(wine.data)

# sns.pairplot(df)
display(df.head())
df.corr()
```

```python
pca = PCA(n_components=len(df.columns))
x_dash = pca.fit_transform(df)
print(pca.explained_variance_ratio_)
```

```python
pca_df = pd.DataFrame(x_dash)
pca_df.corr()
```

```python
import math

def time_to_polar_coordinates(h, m):
    # 
    angle = ((h % 24) + m / 60) * 360 / 24

    # 
    angle_rad = math.radians(angle)

    # (x, y)
    x = math.cos(angle_rad)
    y = math.sin(angle_rad)

    return x, y

# : 1030
x, y = time_to_polar_coordinates(18, 00)
print(f'(x, y): ({x}, {y})')
```
