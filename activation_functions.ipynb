```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-5, 5, 100)
y = nn.Tanh()(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("tanh")
plt.grid()
plt.show()
```

```python
x = torch.linspace(-5, 5, 100)
y = nn.Sigmoid()(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Sigmoid")
plt.grid()
plt.show()
```

```python
x = torch.linspace(-5, 5, 100)
y = nn.ReLU()(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("ReLU")
plt.grid()
plt.show()
```

```python
x = torch.linspace(-5, 5, 100)
y = nn.LeakyReLU()(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("LeakyReLU")
plt.grid()
plt.show()
```

```python
x = torch.linspace(-5, 5, 100)
y = nn.GELU()(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("GELU")
plt.grid()
plt.show()
```

```python
x = torch.linspace(-5, 5, 100)
y = nn.Mish()(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Mish")
plt.grid()
plt.show()
```
