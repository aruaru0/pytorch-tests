## pyplotで日本語

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import random

x = [i for i in range(100)]
y = [i + random.randint(-10, 10) for i in range(100)]

plt.plot(x, y, label="sample")
plt.title("グラフ")
plt.xlabel("x軸")
plt.ylabel("y軸")
plt.legend()
plt.show()
```

## pyplotで日本語（２）

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import random

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for i in range(2):
  x = [i for i in range(100)]
  y = [i + random.randint(-10, 10) for i in range(100)]
  ax.plot(x, y, label=f"sample{i}")
  ax.set_title("グラフ")
  ax.set_ylabel("y軸")
  ax.set_xlabel("x軸")

ax.legend()
plt.show()
```

## seabornで日本語

```python
import seaborn as sns
import random
import numpy as np
import pandas as pd

x = np.arange(0, 100)
y = np.random.randint(0, 100, 100)
df = pd.DataFrame({'軸１': x, '軸２': y})
sns.scatterplot(x='軸１', y='軸２', data=df)
```

## seaborn objectsで日本語

```python
import seaborn as sns
import numpy as np
import pandas as pd
import seaborn.objects as so
import japanize_matplotlib

x = np.arange(0, 100)
x = np.concatenate([x,x])
y = np.random.randint(0, 100, 200)
z = ['A' for _ in range(100)] + ['B' for _ in range(100)]
df = pd.DataFrame({'サンプル１': x, 'サンプル２': y, '結果' : z})

so.Plot(df, x='サンプル１', y = 'サンプル２').add(so.Dot()).theme({'font.family': 'IPAexGothic'})
```
