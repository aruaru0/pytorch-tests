# cyberagent/calm2-7b-chat

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, TextStreamer


model = AutoModelForCausalLM.from_pretrained("cyberagent/calm2-7b-chat",
                                             device_map="auto",
                                             torch_dtype="auto")
tokenizer = AutoTokenizer.from_pretrained("cyberagent/calm2-7b-chat")
streamer = TextStreamer(tokenizer, skip_prompt=True, skip_special_tokens=True)
```

```python
model.eval()
```

```python
prompt = "最大32,000トークンで、約50,000文字ということでトークンがどのように区切られているのか気になりました。"
input_ids = tokenizer(prompt, return_tensors="pt")#, max_length=32, padding=True, truncation=True)

for e in input_ids['input_ids'][0]:
  r = tokenizer.decode(e)
  print(r, end=" | ")
print(f"\n文字数={len(prompt)}, トークン数={len(input_ids['input_ids'][0])}")
```

```python
# prompt = """USER:フィボナッチ数列をN番目まで列挙するプログラムをPythonで作成してください。
# ASSISTANT:
# """,
prompt = """USER: 自転車は車道と歩道のどちらを走行すればよいでしょうか？
ASSISTANT:
"""
input_ids = tokenizer(prompt, return_tensors="pt")#, max_length=32, padding=True, truncation=True)
```

```python
for e in input_ids['input_ids'][0]:
  r = tokenizer.decode(e)
  print(e, r)
```

```python
with torch.no_grad():
  tokens = model.generate(
      **input_ids.to(model.device),
      max_new_tokens=300,
      do_sample=True,
      temperature=0.7,
  )
output = tokenizer.decode(tokens[0], skip_special_tokens=True)
print(output)
```

```python
while True :
  txt = input("入力: ")
  if txt == "quit" : break

  prompt = f"""USER:{txt}
  ASSISTANT:"""
  input_ids = tokenizer.encode(prompt, return_tensors="pt")

  print("アシスタント：", end=" ")
  with torch.no_grad():
    tokens = model.generate(
      input_ids = input_ids.to(model.device),
      max_new_tokens=300,
      do_sample=True,
      temperature=0.7,
      top_k = 50, # op-kサンプリングでは、最も可能性の高いK個の次の単語をフィルタリングし、それらのK個の次の単語のみに確率を再分配
      # top_p = 0.9, # 累積確率が確率pを超える可能性のある最小の単語の集合
      streamer=streamer,
    )
```

```python
prompt = ""
while True :
  txt = input("入力: ")
  if txt == "quit" : break

  prompt += f"USER:{txt}\nASSISTANT:"
  input_ids = tokenizer.encode(prompt, return_tensors="pt")
  print("-"*80)
  print(prompt)
  print("-"*80)

  print("アシスタント：", end=" ")
  with torch.no_grad():
    tokens = model.generate(
      input_ids = input_ids.to(model.device),
      max_new_tokens=300,
      do_sample=True,
      temperature=0.7,
      top_k = 50, # op-kサンプリングでは、最も可能性の高いK個の次の単語をフィルタリングし、それらのK個の次の単語のみに確率を再分配
      # top_p = 0.9, # 累積確率が確率pを超える可能性のある最小の単語の集合
      streamer=streamer,
    )
  output = tokenizer.decode(tokens[0], skip_special_tokens=True)
  # print(output)
  prompt += output.split("ASSISTANT:")[-1] + "<|endoftext|>\n"
```

```python
# パラメータチューニング
# https://huggingface.co/docs/transformers/ja/main_classes/text_generation

# model.generate(
#     input_ids,
#     max_length=512,
#     num_beams=5,
#     no_repeat_ngram_size=4,
#     temperature=temp,
#     num_return_sequences=1,
#     early_stopping=True,
# )
# TL;DR -- there are several generation modes, and not all .generate() parameters are active for a given generation mode.
# In particular, the popular temperature, top_p, and top_k are only active when do_sample=True is also passed.
# Some tasks benefit from do_sample=True, while others do not. Popular tools like ChatGPT operate with sampling.


prompt = ""
while True :
  txt = input("入力: ")
  if txt == "quit" : break
  if txt == "clear" :
    prompt = ""
    continue

  prompt += f"USER:{txt}\nASSISTANT:"
  input_ids = tokenizer.encode(prompt, return_tensors="pt")
  print("-"*80)
  print(prompt)
  print("-"*80)

  print("アシスタント：", end=" ")
  with torch.no_grad():
    tokens = model.generate(
      input_ids = input_ids.to(model.device),
      max_new_tokens=300,
      do_sample=True,
      temperature=0.7,
      no_repeat_ngram_size=2,
      # top_k = 50, # op-kサンプリングでは、最も可能性の高いK個の次の単語をフィルタリングし、それらのK個の次の単語のみに確率を再分配
      top_p = 0.9, # 累積確率が確率pを超える可能性のある最小の単語の集合
      streamer=streamer,
    )
  output = tokenizer.decode(tokens[0], skip_special_tokens=True)
  # print(output)
  prompt += "ASSISTANT:"+output.split("ASSISTANT:")[-1] + "<|endoftext|>\n"
```
