```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
```

```python
from langchain.document_loaders import UnstructuredPDFLoader
```

```python
loader = UnstructuredPDFLoader("kubun.pdf")
test_all = loader.load()
test_all
```

```python
test_all[0].page_content
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=512,
    chunk_overlap=32,
)
texts = text_splitter.split_text(test_all[0].page_content)
# チャンクの確認
print(len(texts))
for text in texts:
    print(text[:10].replace("\n", "\\n"), ":", len(text))
```

```python
# ドキュメントの読み込み
with open("kubun.txt", encoding="utf-8") as f:
    test_all = f.read()
```

```python
# チャンクの分割
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=32,
)
texts = text_splitter.split_text(test_all)
```

```python
# チャンクの確認
print(len(texts))
for text in texts:
    print(text[:10].replace("\n", "\\n"), ":", len(text))
```

```python
%%time
embedding=HuggingFaceEmbeddings(model_name="intfloat/multilingual-e5-large")

index = FAISS.from_texts(
    texts=texts,
    embedding=embedding
)
```

## ロードとセーブ

```python
index.save_local("faiss_index")
```

```python
new_index = FAISS.load_local("faiss_index", embedding)
```

## 検索

```python
docs = index.similarity_search("規約の変更")
docs
```

```python
docs = new_index.similarity_search("規約の変更", k = 1)
docs
```

```python
docs[0].page_content
```

```python
docs = new_index.similarity_search_with_score("規約の変更", k = 5)
docs
```

## Vectorから検索

```python
embedding_vector = embedding.embed_query("規約の変更")
embedding_vector[:10]
```

```python
new_index.similarity_search_by_vector(embedding_vector, k=3)
```
